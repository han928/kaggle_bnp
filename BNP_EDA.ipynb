```python
from __future__ import division
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile, StringIO, requests

%pylab inline
```

```python
z = zipfile.ZipFile('data')
```

```python
df = pd.read_csv(z.open('train.csv'))
```

# EDA

```python
df.head(20)
```

```python
print '%f percentage of data is null for not approved' % (df[df['target']==0]['v131'].isnull().sum()/df[df['target']==0]['v131'].isnull().count())
print '%f percentage of data is null for approved' % (df[df['target']==1]['v131'].isnull().sum()/df[df['target']==1]['v131'].isnull().count())

fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111)
df[df['target']==0]['v131'].hist(bins=100, alpha=0.5, normed = True, ax=ax, label='not approved')
df[df['target']==1]['v131'].hist(bins=100, alpha=0.5, normed = True, ax=ax, label='approved')
ax.legend()
ax.set_title('v131')
ax.set_xlim(-1,21)
```

```python
fig = plt.figure(figsize=(16,9))
ax = fig.add_subplot(111)
(df[df['target']==0]['v129'].value_counts() / df[df['target']==0]['v129'].count()).plot(label = 'not approved')
(df[df['target']==1]['v129'].value_counts() / df[df['target']==1]['v129'].count()).plot(label = 'approved')
ax.legend()
ax.set_title('v129')
ax.set_xlim(-1,15)
print 'not approved', df[df['target']==0]['v129'].value_counts() / df[df['target']==0]['v129'].count()   # could be some difference
print 'approved', df[df['target']==1]['v129'].value_counts() / df[df['target']==1]['v129'].count()   # could be some difference
```
